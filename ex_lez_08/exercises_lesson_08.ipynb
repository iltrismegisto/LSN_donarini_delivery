Nel primo esercizio abbiamo dovuto valutare, attraverso l'algoritmo Variational Monte Carlo, lo stato fondamentale per una particella quantistica unidimensionale confinata nel seguente potenziale:
$$
V(x)=x^4−\frac{5}{2}x^2
$$
Per fare ciò abbiamo dovuto ottimizzare i parametri $\mu $ e $\sigma$ della funzione test:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

Qui si può trovare il grafico di $\langle {\hat H} \rangle_T$ e il suo errore ottenuto con il DataBlocking.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("pot.dat", usecols=(0,1,2), delimiter=';', unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
plt.errorbar(x,f,yerr=error)
plt.xlabel('Blocks')
plt.xlabel("Block Number")
plt.ylabel("Energy")
plt.title('DB Ground State Energy')
plt.grid(True)
plt.show()
```

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 6
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
```

```python
def psi_t(x,mu,sigma):
    return (np.exp(-(x-mu)**2./(2.*sigma**2.))+np.exp(-(x+mu)**2./(2.*sigma**2.)))/np.sqrt(2.*np.sqrt(np.pi)*sigma*(1.+np.exp(-(mu/sigma)**2.)))

mu_opt=0.8 
sigma_opt=0.62

x, y, error = np.loadtxt("hist.dat",usecols=(0,1,2), delimiter=";", unpack='true')
x=np.linspace(-3.,3.,len(y))
plt.errorbar(x, y ,yerr=error,label="Histo")

for i in range(len(y)):
    y[i]=psi_t(x[i],mu_opt,sigma_opt)*psi_t(x[i],mu_opt,sigma_opt)

plt.plot(x,y,label="Test Function with Optimized Parameters")

#Lecture Code
def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-3, 3, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
#plt.figure(figsize=(8,5))
#scale = 0.3
plt.plot(x,(psi[0])**2, label="Numeric")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.title("Squared Modulus of $\psi$ as a Function of x")

plt.grid()
plt.legend()
plt.show()
```

Il grafico appena riportato mostra un istogramma di $|\psi_T(x)|^2$, confrontando l'output Monte Carlo con la curva analitica e la funzione d'onda di prova. 
È possibile notare che i valori ottenuti con il Monte Carlo ha un andamento molto compatibile con i valori ottenuti analiticamente.
