# Es 4.1

In questa esercitazione è stato chiesto di modificare un codice di Dinamica Molecolare. <br>
La prima richiesta è stata quella di implementare la possibilità di far ripartire la simulazione da una situazione precedentemente modificate (diversa dalla situazione cristallina perfetta). <br>
Sono state riscalate le temperature per farle corrispondere alla temperatura di target, lasciando il sistema funzionare per 1000 passi 4 volte, ogni volta partendo dalla configurazione raggiunta nella precedente e facendo corrispondere le velocità al fine di compensare la caduta di energia cinetica associata al raggiungimento dell'equilibrio termico del sistema.<br> 
Tutte le misure sono espresse in unità Lennard-Jones.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16,9]
temp = np.loadtxt("output_temp_generic.dat", usecols=(0), unpack='true')
plt.plot(temp, label="Temperature")
u = np.loadtxt("output_epot_generic.dat", usecols=(0), unpack='true')
plt.plot(u, label="Potential Energy")
k = np.loadtxt("output_ekin_generic.dat", usecols=(0), unpack='true')
plt.plot(k, label="Kinetic Energy")
t = np.loadtxt("output_etot_generic.dat", usecols=(0), unpack='true')
plt.plot(t, label="Total Energy")
plt.xlabel('Measure')
plt.ylabel('Value')
plt.grid(1)
plt.legend()
plt.show()
```

# Es 4.2

La seconda richiesta dell'esercitazione era di aggiungere il data blocking per misurare le proprietà medie e i loro errori statistici. 

```python
block,temp,error = np.loadtxt("ave_epot_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, temp, yerr=error, label="Potential Energy")
plt.title("Es 4.2")
plt.xlabel('Block Number')
plt.ylabel('Potential Energy')
plt.grid(1)
plt.legend()
plt.show()
```

```python
block,temp,error = np.loadtxt("ave_ekin_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, temp, yerr=error, label="Kinetic Energy")
plt.title("Es 4.2")
plt.xlabel('Block Number')
plt.ylabel('Kinetic Energy')
plt.grid(1)
plt.legend()
plt.show()
```

```python
block,temp,error = np.loadtxt("ave_etot_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, temp, yerr=error, label="Total Energy")
plt.title("Es 4.2")
plt.xlabel('Block Number')
plt.ylabel('Total Energy')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```

```python
block,temp,error = np.loadtxt("ave_temp_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, temp, yerr=error, label="Temperature")
plt.title("Es 4.2")
plt.xlabel('Block Number')
plt.ylabel('Temperature')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```
