L'esercitazione numero 9 propone di risolvere il probelma del commesso viggiatore, cioè provare a trovare la sequenza, fra le città che devono esser visitate, di città in modo da percorrere il minor percorso possibile. <br>
Questo è stato ottenuto implementando un algoritmo genetico con i seguenti criteri:
- Un singolo gene consiste in una lista ordinata di città, il costo di un gene è la lunghezza necessaria per visitarle in ordine
- Generare un insieme casuale (3000 elementi) di possibili soluzioni al TSP (chiamato generazione 1)
- Valutare il costo di ogni membro della prima generazione, ordinandoli per questo valore. (il corso è inteso come la quantità da minimizzare, infatti è il percorso fra le varie città).
- Salvare 15 elementi della generazione corrente, scelti a caso con una probabilità ponderata per favorire i membri migliori (giudicati sul loro costo).
- Mutare (con una probabilità del 10%) ogni membro della generazione corrente scambiando una coppia di città, spostando 3 città in posizione casuale, facendo la permutazione di 3 gruppi di 3 città o invertendo 3 città consecutive del gene.
- Incrociare due geni con una probabilità dell'90% e aggiungere il risultato alla generazione successiva.
- Iterate questa procedura (ripartendo dalla valutazione e dall'ordinamento) per 300 generazioni.
- Infine scegliere il miglior gene dell'ultima generazione.

Abbiamo testato l'algoritmo con 32 città poste lungo una circonferenza e all'interno di un quadrato.

La prima distribuzione di città su una circonferenza è stato utilizzato come test per valutare la bontà dell'algoritmo, poiché è scontato che la lunghezza minima del percorso dovrebbe essere $2\pi$. Nella sconda distribuzione ha un problema più interessante le città sono disposte casualmente su un'area quadrata di lato pari a $1$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
e = np.loadtxt("evolution.dat")
x_e = e[:,0]
y_e = e[:,1]

fig, axs = plt.subplots(1, 1, figsize=(15, 10))

plt.plot(x_e,y_e)
plt.title('Salesman path evolution')
plt.xlabel('#Generation')
plt.ylabel('Length')
plt.grid()
plt.show()
```

```python
f = np.loadtxt("first.gene")
x_f = f[:,1]
y_f = f[:,2]

b = np.loadtxt("best.gene")
x_b = b[:,1]
y_b = b[:,2]

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f,y_f)
axs[0].set_title('Salesman first path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b,y_b)
axs[1].set_title('Salesman best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()
```
