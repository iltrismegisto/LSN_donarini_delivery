# Es_51_1 (Uniform Metropolis)

La consegna dell'esercizio 5 è di implementare l'algoritmo Metropolis per campionare gli autostati 1s e 2p degli atomi di idrogeno.<br>
I grafici sottostanti mostrano la distribuzione dei punti nello spazio 3D e il raggio medio degli autostati (in termini di raggio di Bohr $a_0$), calcolati con una distribuzione uniforme e una distribuzione gaussiana.<br>
In ogni run il passo è stato scelto in modo che il tasso di accept/reject sia intorno al 50%.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z =  np.loadtxt("1s_l.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
block, r0, error = np.loadtxt("1s_l_ave.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
plt.errorbar(block,r0,yerr=error)
plt.axhline(y=1.5)
plt.xlabel('Block Number')
plt.ylabel('r')
plt.grid(1)
plt.show()
```

```python
X, Y, Z =  np.loadtxt("2p_l.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0, 30)
plt.show()
```

```python
block, r1, error = np.loadtxt("2p_l_ave.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
plt.errorbar(block,r1,yerr=error)
plt.axhline(y=5)
plt.xlabel('Block Number')
plt.ylabel('r')
plt.grid(1)
plt.show()
```

È possibile osservare che, sia nel caso dell'autostato 1s che per quello 2p, sia nel caso di algoritmo Metropolis campionato con una distribuzione uniforme sia per quello con una distribuzione gaussiana abbiamo dei risultati in linea con quanto teoricamente aspettato. <br>
I valori di aspettazione degli autostati dell'atomo di idrogeno sono stati calcolati a partire dalla funzione d'onda degli autostati di un elettrone in un atomo di idrogeno, espressa in termini di armoniche sferiche e polinomi di Laguerre generalizzati.

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

dove $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm è il raggio di Bohr, $L_{n-l-1}^{2l+1}$ sono i polinomi di Laguerre generalizzati di grado $n-l-1$, $n=1,2,. ...$ è il numero quantico principale, $l=0,1, ..., n-1$ il numero quantico azimutale, $m=-l, -l+1, ..., l-1, l$ il numero quantico magnetico.

Per esempio, la funzione d'onda dello stato fondamentale è:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
mentre uno dei tre stati eccitati di $2p$ è:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
I valori di aspettativa per il raggio risultano essere esattamente:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
