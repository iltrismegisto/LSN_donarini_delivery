# Es_31_1

Lo scopo di questa esercitazione è di valutare il prezzo di opzioni Europee Call & Pull.
Questo è stato fatto utilizzando la teoria di Black-Scholes che presuppone che l'evoluzione temporale del prezzo del bene sia stocastica ed esibisca un movimento geometrico browniano (GBM) con un tasso di interesse costante senza rischio $r$ e una volatilità $\sigma$.
L'esercizio chiede di completare questo calcolo in due modalità: con un campionamento diretto a tempo $T=1$, oppure seguiendo l'evoluzione discreta del prezzo. <br>
I risultati di entrambi i metodi vanno a convergere con il prezzo calcolato analiticamente e, nella maggior parte dei casi, si può vedere che il risultato prodotto dista meno di un $\sigma$ dal valore analitico.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, sum, error = np.loadtxt("direct.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
plt.errorbar(x,sum,yerr=error, marker="x")
plt.title('Sampling Directly - Put')
plt.xlabel('Block Number')
plt.ylabel('<put value>')
plt.grid(True)
plt.axhline(y=5.4595325819072364, color='black')
plt.show()
```

```python
x, sum, error = np.loadtxt("direct.dat", usecols=(0,3,4), delimiter=";", unpack='true')
plt.errorbar(x,sum,yerr=error, marker="x")
plt.rcParams['figure.figsize'] = [16,9]
plt.xlabel('Block Number')
plt.title('Sampling Directly - Call')
plt.ylabel('<call value>')
plt.grid(True)
plt.axhline(y=14.975790778311286, color='black')
plt.show()
```
