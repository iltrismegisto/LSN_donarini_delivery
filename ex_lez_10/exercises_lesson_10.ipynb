# ex_10_1


In questo esercizio abbiamo adattato il codice dell'esercizio precedente (algoritmo genetico) per risolvere il problema del commesso viaggiatore attraverso un algoritmo di Simulated Annealing e in modo da poter eseguire il codice paralelizzandolo. <br>
È stato studiato di nuovo il semplice modello con 32 città intorno ad una circonferenza e poi città disposte randomicamente in un quadrato.
L'algoritmo è stato quindi svolto sia "in serie" che "in parallelo" per varie temperature.
I grafici rappresentano l'andamento della lunghezza del percorso trovato in funzione del numero degli step.
Poi viene presentata la migliore soluzione alla fine del primo step e la migliore soluzione trovata alla fine dell'annealing.
Infine viene fatto un confronto fra il la versione lineare e quella parallelizzata per ciascuna temperatura.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# SERIAL

## T=1

```python
e_c = np.loadtxt("serial/evolution_t1.dat") 
x_e_c = e_c[:,0]
y_e_c = e_c[:,1]

e_s = np.loadtxt("serial/evolution_t1_sq.dat")
x_e_s = e_s[:,0]
y_e_s = e_s[:,1]

fig, axs = plt.subplots(1, 2, figsize=(16, 9))

axs[0].plot(x_e_c,y_e_c)
axs[0].set_title('Salesman path cirle')
axs[0].set_ylabel('Path evolution')
axs[0].set_xlabel('Step')

axs[1].plot(x_e_s,y_e_s)
axs[1].set_title('Salesman path squared')
axs[1].set_ylabel('Path evolution')
axs[1].set_xlabel('Step')

plt.show() 
```

```python
f_c = np.loadtxt("serial/first_t1.gene")
x_f_c = f_c[:,1]
y_f_c= f_c[:,2]

b_c = np.loadtxt("serial/best_t1.gene")
x_b_c = b_c[:,1]
y_b_c = b_c[:,2]

f_s = np.loadtxt("serial/first_t1_sq.gene")
x_f_s = f_s[:,1]
y_f_s= f_s[:,2]

b_s = np.loadtxt("serial/best_t1_sq.gene")
x_b_s_1 = b_s[:,1]
y_b_s_1 = b_s[:,2]


fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f_c,y_f_c)
axs[0].set_title('First path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b_c,y_b_c)
axs[1].set_title('Best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show() 

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f_s,y_f_s)
axs[0].set_title('First path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b_s_1,y_b_s_1)
axs[1].set_title('Best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')


plt.show()
```

# PARALLEL

# T=1

```python
e_p = np.loadtxt("parallel/evolution_t1_sq.dat")
x = e_p[:,0]
y = e_p[:,1]

br = np.loadtxt("parallel/best_rank_evo_t1_sq.dat")
x_br = br[:,0]
y_br = br[:,1]

fig, axs = plt.subplots(2, 1, figsize=(20, 10))

axs[0].plot(x, y)
axs[0].set_title('Salesman path evolution')
axs[0].set_ylabel('Path evolution')
axs[0].set_xlabel('Step')

axs[1].plot(x_br, y_br)
axs[1].set_title('Best cpu node')
axs[1].set_xlabel('Step')
axs[1].set_ylabel('MPI Rank')

plt.show()
```

```python
f_s = np.loadtxt("parallel/first_t1_sq.gene")
x_f_s = f_s[:,1]
y_f_s = f_s[:,2]

b_s = np.loadtxt("parallel/best_t1_sq.gene")
x_b_s = b_s[:,1]
y_b_s = b_s[:,2]

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f_s, y_f_s)
axs[0].set_title('First path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b_s, y_b_s)
axs[1].set_title('Best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_b_s, y_b_s)
axs[0].set_title('Best path - parallel')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b_s_1, y_b_s_1)
axs[1].set_title('Best path - serial')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()

r = 0
for i in range(len(x_b_s)):
    r += ((x_b_s[i]-x_b_s[i-1])**2 + (y_b_s[i]-y_b_s[i-1])**2)**(1/2)
print("The best path cost for the parallel annealing is: " + str(r))
r = 0
for i in range(len(x_b_s_1)):
    r += ((x_b_s_1[i]-x_b_s_1[i-1])**2 + (y_b_s_1[i]-y_b_s_1[i-1])**2)**(1/2)
print("The best path cost for the serial annealing is: " + str(r))
```
